# INTRODUCCIÓN


El negocio bancario se podría definir como un sistema de intermediación financiera, entre aquellos individuos (personas naturales o jurídicas) que poseen excesos de liquidez sobre los cuales buscan algún tipo de rentabilidad, y aquellos individuos que poseen necesidades de liquidez que están dispuestos a pagar algún tipo de interés por la misma. Así las cosas, los bancos se convierten en el instrumento que garantiza que el progreso sea exitoso para las dos partes. 

Es importante indicar que en ese rol de intermediación que realizan los Bancos, se genera un  diferencial o fee entre las tasas de interés que son cobradas a los individuos que buscan liquidez y las tasas de rendimiento que son otorgadas a aquellos individuos que buscan rentabilidad. Es en ese diferencial donde los Bancos generan la mayor parte de sus ingresos, y lo que podría definirse como el core del negocio Bancario. 

La creación de los productos y servicios que son ofrecidos a los clientes, se generan en su mayoría en la identificación de las necesidades de los individuos; ya sea a través de productos estándar o de productos hechos a la medida. Para el caso de los individuos que buscan rentabilizar su liquidez, la mayoría de productos están definidos de acuerdo con el espacio de tiempo y perfil de riesgo del inversionista; un ejemplo de esto son los CDT´s y las inversiones en divisas. Para el caso de los individuos que buscan liquidez, los Bancos crean productos de acuerdo con el tipo de financiación que necesita y el tiempo, entre los mas relevantes. Un ejemplo de esto son las tarjetas de crédito (financiación de bienes de consumo masivo generalmente de corto plazo) y los créditos hipotecarios (financiación de bienes de consumo duradero otorgados a largo plazo). 
La promoción de los servicios bancarios se realiza generando factores diferenciadores con respecto a las demás entidades Bancarias; ya sean en términos de tasas, experiencia para los consumidores o beneficios adicionales.

Definimos una función para analizar las variables en 2 dimensiones:

```python
def plot_two_classes(X, y, subplot=False, size=(10, 10)):
    # Plot the two classes
    if subplot == False:
        fig, subplot = plt.subplots(nrows=1, ncols=1, figsize=size)
        
    subplot.scatter(X[y==0, 0], X[y==0, 1], label="Class #0", 
                    alpha=0.5, s=70)
    subplot.scatter(X[y==1, 0], X[y==1, 1], label="Class #1", 
                    alpha=0.5, s=70)
    subplot.legend()
    return subplot
```

Graficando la distribución de los datos en 2 componentes principales:

```python
pca = PCA(n_components = 2)
x_vis = pca.fit_transform(X)
plot_two_classes(x_vis, y)
```

```python
n_samples = y.shape[0]
n_samples_0 = (y == 0).sum()
n_samples_1 = (y == 1).sum()
n_samples_1 / n_samples
```

Efectivamente hay un desbalanceo en la información, con una razon de 0.1325 de la clase minoritaria respecto a la mayoritaria.

## Balanceo de la información

Se realiza un balanceo utilizando variables sintéticas:

```python
def SMOTE(X, y, target_percentage=0.5, k=5, seed=None):
    
    # New samples
    n_samples_1_new =  int(-target_percentage * n_samples_0 / (target_percentage- 1) - n_samples_1)
    
    # A matrix to store the synthetic samples
    new = np.zeros((n_samples_1_new, x_vis.shape[1]))
    
    # Create seeds
    np.random.seed(seed)
    seeds = np.random.randint(1, 1000000, 3)
    
    # Select examples to use as base
    np.random.seed(seeds[0])
    sel_ = np.random.choice(y[y==1].shape[0], n_samples_1_new)
    
    # Define random seeds (2 per example)
    np.random.seed(seeds[1])
    nn__ = np.random.choice(k, n_samples_1_new)
    np.random.seed(seeds[2])
    steps = np.random.uniform(size=n_samples_1_new)  

    # For each selected examples create one synthetic case
    for i, sel in enumerate(sel_):
        # Select neighbor
        nn_ = nn__[i]
        step = steps[i]
        # Create new sample
        new[i, :] = X[y==1][sel] - step * (X[y==1][sel] - X[y==1][nn_])
    
    X = np.vstack((X, new))
    y = np.append(y, np.ones(n_samples_1_new))
    
    return X, y
```

```python
for target_percentage in [0.25, 0.5]:
    for k in [5, 15]:
        X_u, y_u = SMOTE(x_vis, y, target_percentage, k, seed=3)
        print('Target percentage', target_percentage, 'k ', k)
        print('y.shape = ',y_u.shape[0], 'y.mean() = ', y_u.mean())
        plot_two_classes(X_u, y_u, size=(5, 5))
        plt.show()    
```

Con los datos balanceados, se realiza un modelo de clasificación:

```python
y_u_d = pd.DataFrame(data=y_u)
X_u_d = pd.DataFrame(data=X_u)
X_train, X_test, y_train, y_test = train_test_split(X_u_d, y_u_d, test_size=0.3, random_state=42)
```

```python
clf = RandomForestClassifier(max_depth=5, n_estimators = 5,max_features=2,random_state=123)
X= clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print(metrics.accuracy_score(y_pred, y_test))
print(metrics.f1_score(y_pred, y_test))
```
